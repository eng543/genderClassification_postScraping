```python
# read in tweets from Erin's DBs
t1 = pd.read_csv("scraped_tweets1.csv")
t2 = pd.read_csv("scraped_tweets2.csv")
t3 = pd.read_csv("scraped_tweets3.csv")
t4 = pd.read_csv("scraped_tweets4.csv")
```

```python
frames2 = [tweets_erinDB_filtered, t5]
tweets = pd.concat(frames2)
```

```python
# remove non-english tweets, reset indices
tweets_eng = tweets[tweets['language'].isin(['en', 'und'])]
tweets_eng = tweets_eng.reset_index(drop = True)
```

```python
# filter users who have less than 100 tweets
tweets_eng['count'] = 1
tweet_count_grouped = tweets_eng.groupby('query')
tweet_count = tweet_count_grouped['count'].agg([np.sum])
tweet_count['keep'] = tweet_count['sum'] >= 100
tweet_keep = tweet_count[tweet_count['keep'] == True]
users_tweet_keep = list(tweet_keep.index)
tweets_eng_keep = tweets_eng[tweets_eng['query'].isin(users_tweet_keep)]

#len(tweets_eng_keep), len(tweets_eng)
```

```python
# filter users who have 3x as many friends as followers (potential bots)
friend_count_grouped = tweets_eng_keep.groupby('query')
followers_count = friend_count_grouped['num_followers'].agg([np.mean])
friends_count = friend_count_grouped['num_friends'].agg([np.mean])
friends_count['ratio'] = friends_count['mean']/friends_count['mean']
friends_count['keep'] = friends_count['ratio'] <= 3
ratio_keep = friends_count[friends_count['keep'] == True]
users_ratio_keep = list(ratio_keep.index)
tweets_trimmed = tweets_eng_keep[tweets_eng_keep['query'].isin(users_ratio_keep)]
#len(tweets_trimmed), len(tweets_eng_keep), len(tweets_eng)
tweets_trimmed = tweets_trimmed.reset_index(drop = True)
```

```python
# deal with NaN
tweets_trimmed['entities_hashtags'] = tweets_trimmed['entities_hashtags'].fillna(0)
tweets_trimmed['entities_urls'] = tweets_trimmed['entities_urls'].fillna(0)
```

```python
# expressions to remove punctuation, urls, hashtags, @mentions, numbers, emoticons
punctuation = set(string.punctuation)
otherPunct = ["''", "' '", "", " ", "...", "'", "' ", " '"]
url = re.compile(r'^http[s]?://(?:[a-z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-f][0-9a-f]))+')
hashtags = re.compile(r'^(?:\#+[\w_]+[\w\'_\-]*[\w_]+)')
mentions = re.compile(r'^(?:@[\w_]+)')
numbers = re.compile(r'(\d+)\D*(\d*)\D*(\d*)\D*(\d*)') # includes phone numbers
smiley = re.compile(r'[:=;\|\)\(\[\]\{\}][oO\-]?[D\)\]\(\[/\\OpP\|\{\}:]')
remove_regex = [url, mentions, numbers, smiley] #, hashtags
remove_regex_with_hash = [url, hashtags, mentions, numbers, smiley]
```

```python
tweets_subset1 = tweets_trimmed[tweets_trimmed['query'] == "ForeverMoreVids"]
tweets_subset2 = tweets_trimmed[tweets_trimmed['query'] == "brianjdixon"]
subsets = [tweets_subset1, tweets_subset2]
tweets_subset = pd.concat(subsets)
tweets_subset = tweets_subset.reset_index(drop = True)
```

```python
tknzr = TweetTokenizer(strip_handles = True, reduce_len = True)
```

```python
# hashtags not included as unigrams, bi/trigrams including hashtags removed
#ngramDict = {} # includes POS tags
ngramDict = Counter()

#count = 0

user_count = 1
for user in set(tweets_subset['query']):
    print user_count
    user_sub = tweets_subset[tweets_subset['query'] == user]
    tweets = list(set(user_sub['content']))
    for tweet in tweets:
        terms_all = []
        terms = tknzr.tokenize(tweet.lower())
        terms_all.append((term.encode('ascii', 'ignore') for term in terms))
        for term in terms:
            if (not any(rr.search(term) for rr in remove_regex)) and (term not in punctuation) and not (term.startswith('www')) and (term not in otherPunct):
#                if term.encode('ascii', 'ignore') not in ngramDict:
#                    ngramDict[term.encode('ascii', 'ignore')] = count
#                    count += 1
                ngramDict[term.encode('ascii', 'ignore')] += 1

                if term.encode('ascii', 'ignore') != "":
                    pos = pos_tag([term.encode('ascii', 'ignore')])
#                    if pos[0][1] not in ngramDict:
                        #ngramDict[pos[0][1]] = count
                        #count += 1
                    ngramDict[pos[0][1]] += 1
        
        for tweet in terms_all:
            for term in list(bigrams(tweet)):
                if ((not any(rr.search(term[0]) for rr in remove_regex_with_hash)) and (term[0] not in punctuation) and not (term[0].startswith('www')) and (term[0] not in otherPunct)) and ((not any(rr.search(term[1]) for rr in remove_regex_with_hash)) and (term[1] not in punctuation) and not (term[1].startswith('www')) and (term[1] not in otherPunct)):
                    #ngramDict[term] = count
                    #count += 1
                    ngramDict[str(term)] += 1

        for tweet in terms_all:
            for term in list(trigrams(tweet)):
                if ((not any(rr.search(term[0]) for rr in remove_regex_with_hash)) and (term[0] not in punctuation) and not (term[0].startswith('www')) and (term[0] not in otherPunct)) and ((not any(rr.search(term[1]) for rr in remove_regex_with_hash)) and (term[1] not in punctuation) and not (term[1].startswith('www')) and (term[1] not in otherPunct)) and ((not any(rr.search(term[2]) for rr in remove_regex_with_hash)) and (term[2] not in punctuation) and not (term[2].startswith('www')) and (term[2] not in otherPunct)):
                    #ngramDict[term] = count
                    #count += 1
                    ngramDict[str(term)] += 1
    user_count += 1

#print max(ngramDict.values()), count
ngramDict['function_words'] += 1# = count
ngramDict['content_words'] += 1 #= count + 1
```

```python
ngram_map = {x:y for y,x in enumerate(ngramDict.keys())}
```

```python
# read back in in case of kernel death
with open('ngram_map.txt', 'r') as rf:
    ngram_map = json.loads(rf.read())
```

```python
user_feature_matrix = np.zeros((len(set(tweets_trimmed['query'])), len(ngram_map)))

function_tags = ['CC', 'DT', 'EX', 'IN', 'MD', 'PDT', 'PRP', 'PRP$', 'RP', 'TO', 'WDT', 'WP', 'WP$', 'WRB']

user_index = 0
for user in list(set(tweets_trimmed['query'])):
    print user, user_index
    user_sub = tweets_trimmed[tweets_trimmed['query'] == user]
    user_dict = Counter()
    user_pos = Counter()
    user_class = Counter()
    tweets = list(set(user_sub['content']))

#        user_feature_list = [0] * len(ngram_map)
    for tweet in tweets:
        terms_all = []
        terms = tknzr.tokenize(tweet.lower())
        terms_all.append((term.encode('ascii', 'ignore') for term in terms))

        for uni in terms:
            if (not any(rr.search(uni) for rr in remove_regex)) and not (uni.startswith('www')) and (uni not in otherPunct):
                if (uni not in punctuation):
                    user_dict[uni.encode('ascii', 'ignore')] += 1

                    if uni.encode('ascii', 'ignore') != "":
                        pos = pos_tag([uni.encode('ascii', 'ignore')])
                        user_pos[pos[0][1]] += 1

                        if pos[0][1] in function_tags:
                            user_class['function_words'] += 1
                        else:
                            user_class['content_words'] += 1

        for tweet in terms_all:
            for bi in list(bigrams(tweet)):
                if ((not any(rr.search(bi[0]) for rr in remove_regex_with_hash)) and (bi[0] not in punctuation) and not (bi[0].startswith('www')) and (bi[0] not in otherPunct)) and ((not any(rr.search(bi[1]) for rr in remove_regex_with_hash)) and (bi[1] not in punctuation) and not (bi[1].startswith('www')) and (bi[1] not in otherPunct)):
                    user_dict[str(bi)] += 1

        for tweet in terms_all:
            for tri in list(trigrams(tweet)):
                if ((not any(rr.search(tri[0]) for rr in remove_regex_with_hash)) and (tri[0] not in punctuation) and not (tri[0].startswith('www')) and (tri[0] not in otherPunct)) and ((not any(rr.search(tri[1]) for rr in remove_regex_with_hash)) and (tri[1] not in punctuation) and not (tri[1].startswith('www')) and (tri[1] not in otherPunct)) and ((not any(rr.search(tri[2]) for rr in remove_regex_with_hash)) and (tri[2] not in punctuation) and not (tri[2].startswith('www')) and (tri[2] not in otherPunct)):
                    user_dict[str(tri)] += 1

#    print user_dict
    for ngram in user_dict.keys():
        ngram_index = ngram_map[ngram]
        user_feature_matrix[user_index][ngram_index] += 1
#            user_feature_list[ngram_index] += 1

    for pos in user_pos.keys():
        pos_index = ngram_map[pos]
        # pos feature = proportion that part of speech used out of all pos
        user_feature_matrix[user_index][pos_index] += user_pos[pos]/float(sum(user_pos.values()))
#            user_feature_list[ngram_index] += user_pos[pos]/float(sum(user_pos.values()))

    for word_class in user_class.keys():
        word_class_index = ngram_map[word_class]
        user_feature_matrix[user_index][word_class_index] += user_class[word_class]/float(sum(user_class.values()))        
#            user_feature_list[ngram_index] += user_class[pos]/float(sum(user_class.values()))

    nonzero = user_feature_matrix[user_index].nonzero()
#    nonzero = zip(x, y)
    with open('nonzero_features.txt', 'a') as wf:
        for feature_index in nonzero[0]:
            wf.write(str(user_index) + " " + str(feature_index) + " " + str(user_feature_matrix[user_index, feature_index]) + "\n")
#            wf.write('{:d} {:d} {:g}\n'.format(user_index, feature_index, user_feature_matrix[user_index, feature_index]))

#        for j in len(user_feature_list):
#            wf.write(user_index + "\t" + j + "\t" + user_feature_list[j] + "\n")

    user_index += 1
```

```python
# read in nonzero_features.txt to refill matrix in event of kernel death
with open('nonzero_features.txt', 'r') as rf:
    nonzeros = rf.readlines()

data = []
for line in nonzeros:
    yep = line.split(' ')
    sub_data = (yep[0], yep[1], yep[2][:-1])
    data.append(sub_data)

user_feature_matrix = np.zeros((len(set(tweets_trimmed['query'])), len(ngram_map)))

for dat in data:
    user_feature_matrix[int(dat[0])][int(dat[1])] += float(dat[2])
```

```python
# filter features
# remove features not used by at least 1% of the users
# sum down the "columns"

min_sum = 0.01 * user_feature_matrix.shape[0]
transposed = user_feature_matrix.T
delete_set = []

for i in range(transposed.shape[0]):
    feature_sum = np.sum(transposed[i], 0)
    if feature_sum < min_sum and feature_sum%1 == 0:
        delete_set.append(i)

trimmed_transposed = np.delete(transposed, delete_set, 0)  
trimmed = trimmed_transposed.T
```
